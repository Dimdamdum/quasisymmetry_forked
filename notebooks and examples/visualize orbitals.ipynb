```python
import pyscf
from pyscf.tools import cubegen
```

```python
molpath = "../hamiltonians/N2_bond1.1000.chk"
mol = pyscf.lib.chkfile.load_mol(molpath)
mf = pyscf.scf.RHF(mol)
mf.update_from_chk(molpath)
```

```python
cubegen.orbital(mol, 'N2_1_1_mo_4.cube', mf.mo_coeff[:, 4])
```

```python
cubegen.orbital(mol, 'mo_3.cube', mf.mo_coeff[:,3])
```

```python
cubegen.orbital(mol, 'mo_6.cube', mf.mo_coeff[:, 6])
```
